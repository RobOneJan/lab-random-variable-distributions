## LOGtransform

```python
def log_transform_clean(x):
    if np.isfinite(x) and x!=0:
        return np.log(x)
    else:
        return np.NAN
```

```python
for i in numeric.columns:
    numeric[i] = list(map(log_transform_clean, numeric[i]))
    numeric[i] = numeric[i].fillna(np.mean(numeric[i]))
```

```python
for x in numeric.columns:
    sns.distplot(numeric[x], bins=50)
    plt.show()
```
